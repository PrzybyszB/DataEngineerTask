### Load data to df

```python
import pandas as pd

df = pd.read_csv('Data/data.csv')
df
```

### Count total_value for each product


```python
df['total_value'] = df['quantity'] * df['price']

df
```

### Filter value thats above 1000

```python
df = df.loc[df['total_value'] > 1000]
df = df.sort_values('total_value', ascending=False)
df
```

### Import to PostgreSQL


```python
from sqlalchemy import create_engine
from dotenv import load_dotenv
import os

load_dotenv()

user = os.getenv("DB_USER")
password = os.getenv("DB_PASSWORD")
host = os.getenv("DB_HOST")
port = os.getenv("DB_PORT")
db_name = os.getenv("DB_NAME")


engine = create_engine(f"postgresql://{user}:{password}@{host}:{port}/{db_name}")
df.to_sql('high_value_orders', engine, if_exists='replace', index=False)
```
